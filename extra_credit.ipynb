* Plot the timing results for each query. Represent execution times  (index vs. no index) on the y-axis and num. of records on the x-axis

```python
import matplotlib.pyplot as plt
```

```python
# Query data to graph
point_query_without=(38874, 63198, 85032)
point_query_index=(16358, 20545, 13010)
range_query_without=(35132, 55304, 81002)
range_query_index=(11388, 11245, 8451)
accounts=(50000, 100000, 150000)
```

```python
# Organize by index vs no index (without) and range vs point query
plt.figure(figsize=(9, 6))
plt.plot(accounts, point_query_without, linestyle='-', color='black', label='Point Query - Without')
plt.plot(accounts, point_query_index, linestyle='-', color='pink', label='Point Query - Index')
plt.plot(accounts, range_query_without, linestyle='-', color='blue', label='Range Query - Without')
plt.plot(accounts, range_query_index, linestyle='-', color='green', label='Range Query - Index')

# Graph Title and Axis

plt.title('Query Run Times: Index vs No index')
plt.xlabel('Num. Records')
plt.ylabel('Execution Times(MS)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
